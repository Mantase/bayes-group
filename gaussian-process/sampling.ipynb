```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
points = [
    [-5, 1],
    [1, -5],
    [3, -2],
    [-3, 2],
    [-1, 0.5],
    [6, 1],
    [7, 4],
    [0, -2],
    [5, 0.5],
]
points = np.array(points)
plt.figure()
plt.plot(points[:, 0], points[:, 1], 'o')
```

```python
def kernel(x1, x2):
    return np.exp(-np.linalg.norm(x1 - x2))/1.0
```

```python
cov_mat = [[kernel(x1, x2) for x2, _ in points] for x1, _ in points]
```

```python
mn = stats.multivariate_normal(mean=np.zeros(points.shape[0]), cov=np.array(cov_mat))
```

```python
plt.figure()
plt.plot(points[:, 0], [mn.pdf(p[0]) for p in points], 'o')
```
